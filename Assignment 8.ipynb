```python
import math
from numpy import *

time = linspace(0,10,20)
rate = 15
prin = 5000
i = prin+prin*rate*time/100
j = []
for x in time:
    j.append(prin*math.pow((1+rate/1200), 12*x))

plt.plot(time,i,'blue')
plt.plot(time,j,'red')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.title("Compare simple andcompound interest")
plt.legend(['Simple interest', 'Compound interest'])

plt.show()

```
